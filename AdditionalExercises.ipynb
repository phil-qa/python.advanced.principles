## Advanced variables - namedTuple
A useful processing task for machine learning or positional elements is an understanding of where two objects appear in the spatial domain, these points can be identified in a 2 dimensional context as two integers within a 2d space x, y   
using namedtuple to define a point x and y, create functions to calculate the distance between two points given that   

$distance=\sqrt{(x_2 - x_1)^2+(y_2 -y_1)^2}$   
Addistionally create a function that can modify the positions of a point by moving either x or y in either direction and report the new position

```python
from collections import namedtuple
import math

# Define the Point2D namedtuple
Point = namedtuple('Point', ['x', 'y'])

# Function to create a new Point2D
def create_point(x, y):
    return Point(x, y)

# Function to calculate the distance between two Point2D objects
def distance(point1, point2):
    return math.sqrt((point2.x - point1.x) ** 2 + (point2.y - point1.y) ** 2)

# Function to move a Point2D by dx and dy
def move_point(point, dx, dy):
    return Point(point.x + dx, point.y + dy)

# test the assertions
if __name__ == "__main__":
    # Create two points
    p1 = create_point(1, 2)
    p2 = create_point(4, 6)

    # Calculate the distance between p1 and p2
    print(f"Distance between p1 and p2: {distance(p1, p2)}")

    # Move point p1 by 3 units in the x direction and 4 units in the y direction
    p3 = move_point(p1, 3, 4)
    print(f"New position of p1 after moving: ({p3.x}, {p3.y})")
```

## Advanced Variables Decorator using functool wraps 
Exercise: Implementing a Memoization Decorator with functools.wraps  
In Python, a memoization decorator is a function that optimizes another function by storing previously calculated results and reusing them for subsequent calls with the same arguments. This technique improves the performance of functions that involve repetitive calculations with the same inputs.   



Write a memoization decorator that caches the results of function calls to improve efficiency, and use functools.wraps to preserve the original function’s metadata.   

A fibbonaci functio nis one such operation that could so with this. Given a basic  fibonnacci calculator to the nth number optimise the function using an appropriate decorator

```python
def fibonacci(n):
    """Returns the nth Fibonacci number."""
    if n in (0, 1):
        return n
    return fibonacci(n-1) + fibonacci(n-2)
result = fibonacci(35)
print(result)
```

```python
import functools

def memoize(func):
    cache = {}
    
    @functools.wraps(func)
    def wrapper(*args):
        if args in cache:
            return cache[args]
        result = func(*args)
        cache[args] = result
        return result
    
    return wrapper

@memoize
def fibonacci(n):
    """Returns the nth Fibonacci number."""
    if n in (0, 1):
        return n
    return fibonacci(n-1) + fibonacci(n-2)

# Call the decorated function
result = fibonacci(35)

# Print the result and check the metadata
print(f"Result: {result}")
print(f"Function name: {fibonacci.__name__}")
print(f"Function docstring: {fibonacci.__doc__}")
```

## union 
This exercise shows how we can use the union to provide appropriate types into the code 
Using Union for Flexible Type Hints
You are tasked with creating a function safe_sum that can take a list of numbers, where the numbers can be either integers (int) or floating-point numbers (float). The function should return the sum of these numbers as a floating-point number.   As a part of this process you should implement a negative test case 

```python
# Import necessary type hinting tools
from typing import Union, List

def safe_sum(numbers: List[Union[int, float]]) -> float:
    ''' This function takes either integer or float and returns the sum '''
    total: float = 0.0
    for num in numbers:
        total += num
    return total

# Test the function with a list of mixed integer and floating-point numbers
mixed_numbers = [1, 2.5, 3, 4.75]
result = safe_sum(mixed_numbers)
print(result)  # Expected output: 11.25
```

### Trying the Factory 
One possible use case for the Factory pattern is to create objects without specifying the exact class of object that will be created. 

The Factory pattern is particularly useful in scenarios where the creation process is complex, involves multiple steps, or requires specific conditions to be met.

Problem statement:   
Creating Different Types of Bank Accounts
You are developing a banking application that allows users to open different types of bank accounts, such as Checking, Savings, and Business accounts. Each type of account has different properties and behaviors, but they all share some common functionality.  
Create an appropriate factory pattern that allows for the creation of an account based on a user input. Checking should be a standard account with an interest rate of 2%, savings account should have a higer rate of 5% but if the balance is below £3000 then it should be 2%, business has a 2.5% fixed interest.


```python
from abc import ABC, abstractmethod

class Account(ABC):
    @abstractmethod
    def account_type(self):
        pass

    @abstractmethod
    def interest_rate(self):
        pass

# Create concrete classes
class CheckingAccount(Account):
    def account_type(self):
        return "Checking Account"

    def interest_rate(self):
        return 0.01

class SavingsAccount(Account):
    def account_type(self):
        return "Savings Account"

    def interest_rate(self):
        return 0.03

class BusinessAccount(Account):
    def account_type(self):
        return "Business Account"

    def interest_rate(self):
        return 0.02
```

```python
from flask import Flask, jsonify

app = Flask(__name__)

# Sample stock data
stocks = {
    'AAPL': 150.12,
    'GOOGL': 2750.45,
    'MSFT': 300.22,
    'AMZN': 3400.45
}

@app.route('/stock/<symbol>', methods=['GET'])
def get_stock_price(symbol):
    try:
        symbol = symbol.upper()
        if symbol in stocks:
            return jsonify({'symbol': symbol, 'price': stocks[symbol]})
        else:
            return jsonify({'error': 'Stock not found'}), 404
    except Exception as e:
        return jsonify({'error': str(e)}), 500

if __name__ == '__main__':
    try:
        app.run(debug=True, host='0.0.0.0', port=5000)
    except Exception as e:
        print(f"Failed to start the server: {e}")
```

```python
import requests  # Import the requests module to handle HTTP requests
from bs4 import BeautifulSoup  # Import BeautifulSoup for parsing HTML
import urllib.robotparser  # Import robotparser to handle robots.txt rules
from urllib.parse import urlparse, urljoin  # Import urlparse and urljoin for URL manipulation

# Function to check if a URL is allowed to be scraped according to robots.txt
def is_allowed(url, user_agent='*'):
    # Parse the URL to get the base URL
    parsed_url = urlparse(url)
    base_url = f'{parsed_url.scheme}://{parsed_url.netloc}'
    robots_url = urljoin(base_url, 'robots.txt')
    
    # Parse robots.txt
    rp = urllib.robotparser.RobotFileParser()
    rp.set_url(robots_url)
    rp.read()
    
    # Check if the URL is allowed to be accessed
    return rp.can_fetch(user_agent, url)

# Function to fetch and parse a webpage
def fetch_page(url):
    # Check if the URL is allowed to be scraped
    if not is_allowed(url):
        print(f'Scraping not allowed for {url}')
        return None
    
    try:
        # Send a GET request to the URL
        response = requests.get(url)
        # Check if the request was successful
        if response.status_code == 200:
            print(f'Successfully fetched {url}')
            # Parse the HTML content of the page
            soup = BeautifulSoup(response.content, 'html.parser')
            return soup
        else:
            print(f'Failed to fetch {url} with status code {response.status_code}')
    except Exception as e:
        print(f'Exception occurred while fetching {url}: {e}')
    return None

# Function to extract all links from a webpage
def extract_links(soup, base_url):
    links = []
    if soup:
        # Find all anchor tags with href attribute
        for link in soup.find_all('a', href=True):
            # Resolve relative URLs
            full_url = urljoin(base_url, link['href'])
            links.append(full_url)
    return links

# Function to scrape a list of URLs sequentially
def scrape_urls(urls):
    results = []
    for url in urls:
        page = fetch_page(url)
        if page:
            results.append(page)
    return results

# Main function to start the web scraper
def main():
    start_url = 'https://example.com'  # Replace with the URL you want to start scraping from
    #start_url = 'https://google.com'  # Replace with the URL you want to start scraping from
    soup = fetch_page(start_url)
    if not soup:
        return
    
    # Extract links from the start page
    links = extract_links(soup, start_url)
    # Scrape the extracted links
    pages = scrape_urls(links)
    
    # Optionally, you can further process the scraped pages
    for page in pages:
        # Example: print the title of each page
        if page:
            title = page.find('title').get_text()
            print(f'Page title: {title}')

if __name__ == '__main__':
    main()
```

```python
import asyncio
import aiohttp
from bs4 import BeautifulSoup
import urllib.robotparser
from urllib.parse import urlparse, urljoin

# Function to check if a URL is allowed to be scraped according to robots.txt
async def is_allowed(url, user_agent='*'):
    parsed_url = urlparse(url)
    base_url = f'{parsed_url.scheme}://{parsed_url.netloc}'
    robots_url = urljoin(base_url, 'robots.txt')
    
    async with aiohttp.ClientSession() as session:
        async with session.get(robots_url) as response:
            text = await response.text()
            rp = urllib.robotparser.RobotFileParser()
            rp.parse(text.splitlines())
            return rp.can_fetch(user_agent, url)

# Function to fetch and parse a webpage asynchronously
async def fetch_page(url):
    if not await is_allowed(url):
        print(f'Scraping not allowed for {url}')
        return None
    
    try:
        async with aiohttp.ClientSession() as session:
            async with session.get(url) as response:
                if response.status == 200:
                    print(f'Successfully fetched {url}')
                    html = await response.text()
                    soup = BeautifulSoup(html, 'html.parser')
                    return soup
                else:
                    print(f'Failed to fetch {url} with status code {response.status}')
    except Exception as e:
        print(f'Exception occurred while fetching {url}: {e}')
    return None

# Function to extract all links from a webpage
def extract_links(soup, base_url):
    links = []
    if soup:
        for link in soup.find_all('a', href=True):
            full_url = urljoin(base_url, link['href'])
            links.append(full_url)
    return links

# Asynchronous function to scrape a list of URLs
async def scrape_urls(urls):
    tasks = []
    for url in urls:
        tasks.append(fetch_page(url))
    return await asyncio.gather(*tasks)

# Main coroutine to start the web scraper
async def main():
    start_url = 'https://example.com'  # Replace with the URL you want to start scraping from
    #start_url = 'https://google.com'  # Replace with the URL you want to start scraping from
    soup = await fetch_page(start_url)
    if not soup:
        return
    
    links = extract_links(soup, start_url)
    pages = await scrape_urls(links)
    
    for page in pages:
        if page:
            title = page.find('title').get_text()
            print(f'Page title: {title}')

if __name__ == '__main__':
    asyncio.run(main())
```
