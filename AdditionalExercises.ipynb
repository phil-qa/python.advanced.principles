## Advanced variables - namedTuple
A useful processing task for machine learning or positional elements is an understanding of where two objects appear in the spatial domain, these points can be identified in a 2 dimensional context as two integers within a 2d space x, y   
using namedtuple to define a point x and y, create functions to calculate the distance between two points given that   

$distance=\sqrt{(x_2 - x_1)^2+(y_2 -y_1)^2}$   
Addistionally create a function that can modify the positions of a point by moving either x or y in either direction and report the new position

```python
from collections import namedtuple
import math

# Define the Point2D namedtuple
Point = namedtuple('Point', ['x', 'y'])

# Function to create a new Point2D
def create_point(x, y):
    return Point(x, y)

# Function to calculate the distance between two Point2D objects
def distance(point1, point2):
    return math.sqrt((point2.x - point1.x) ** 2 + (point2.y - point1.y) ** 2)

# Function to move a Point2D by dx and dy
def move_point(point, dx, dy):
    return Point(point.x + dx, point.y + dy)

# test the assertions
if __name__ == "__main__":
    # Create two points
    p1 = create_point(1, 2)
    p2 = create_point(4, 6)

    # Calculate the distance between p1 and p2
    print(f"Distance between p1 and p2: {distance(p1, p2)}")

    # Move point p1 by 3 units in the x direction and 4 units in the y direction
    p3 = move_point(p1, 3, 4)
    print(f"New position of p1 after moving: ({p3.x}, {p3.y})")
```

## Advanced Variables Decorator using functool wraps 
Exercise: Implementing a Memoization Decorator with functools.wraps  
In Python, a memoization decorator is a function that optimizes another function by storing previously calculated results and reusing them for subsequent calls with the same arguments. This technique improves the performance of functions that involve repetitive calculations with the same inputs.   



Write a memoization decorator that caches the results of function calls to improve efficiency, and use functools.wraps to preserve the original function’s metadata.   

A fibbonaci functio nis one such operation that could so with this. Given a basic  fibonnacci calculator to the nth number optimise the function using an appropriate decorator

```python
def fibonacci(n):
    """Returns the nth Fibonacci number."""
    if n in (0, 1):
        return n
    return fibonacci(n-1) + fibonacci(n-2)
result = fibonacci(35)
print(result)
```

```python
import functools

def memoize(func):
    cache = {}
    
    @functools.wraps(func)
    def wrapper(*args):
        if args in cache:
            return cache[args]
        result = func(*args)
        cache[args] = result
        return result
    
    return wrapper

@memoize
def fibonacci(n):
    """Returns the nth Fibonacci number."""
    if n in (0, 1):
        return n
    return fibonacci(n-1) + fibonacci(n-2)

# Call the decorated function
result = fibonacci(35)

# Print the result and check the metadata
print(f"Result: {result}")
print(f"Function name: {fibonacci.__name__}")
print(f"Function docstring: {fibonacci.__doc__}")
```

## union 
This exercise shows how we can use the union to provide appropriate types into the code 
Using Union for Flexible Type Hints
You are tasked with creating a function safe_sum that can take a list of numbers, where the numbers can be either integers (int) or floating-point numbers (float). The function should return the sum of these numbers as a floating-point number.   As a part of this process you should implement a negative test case 

```python
# Import necessary type hinting tools
from typing import Union, List

def safe_sum(numbers: List[Union[int, float]]) -> float:
    ''' This function takes either integer or float and returns the sum '''
    total: float = 0.0
    for num in numbers:
        total += num
    return total

# Test the function with a list of mixed integer and floating-point numbers
mixed_numbers = [1, 2.5, 3, 4.75]
result = safe_sum(mixed_numbers)
print(result)  # Expected output: 11.25
```

```python
from flask import Flask, jsonify

app = Flask(__name__)

# Sample stock data
stocks = {
    'AAPL': 150.12,
    'GOOGL': 2750.45,
    'MSFT': 300.22,
    'AMZN': 3400.45
}

@app.route('/stock/<symbol>', methods=['GET'])
def get_stock_price(symbol):
    try:
        symbol = symbol.upper()
        if symbol in stocks:
            return jsonify({'symbol': symbol, 'price': stocks[symbol]})
        else:
            return jsonify({'error': 'Stock not found'}), 404
    except Exception as e:
        return jsonify({'error': str(e)}), 500

if __name__ == '__main__':
    try:
        app.run(debug=True, host='0.0.0.0', port=5000)
    except Exception as e:
        print(f"Failed to start the server: {e}")
```

```python
#client code 
import requests

def get_stock_price(symbol: str) -> None:
    try:
        response = requests.get(f'http://<instructor-ip>:5000/stock/{symbol}')
        if response.status_code == 200:
            data = response.json()
            print(f"The price of {data['symbol']} is {data['price']}")
        elif response.status_code == 404:
            print(f"Stock {symbol} not found.")
        else:
            print(f"Error: {response.status_code}")
    except requests.RequestException as e:
        print(f"Network error: {e}")

def main():
    while True:
        symbol = input("Enter the stock symbol (or 'exit' to quit): ")
        if symbol.lower() == 'exit':
            break
        get_stock_price(symbol)

if __name__ == '__main__':
    main()
```
