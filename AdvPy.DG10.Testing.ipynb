# Testing 
For testing, we have several tools available in the standard library.
The profiler tool will show where your code is chewing-up the time,
which functions are being called, and how long it spends in each.
The trace module does a similar thing, but in more detail. It tells
you which lines of code are executed (and which are not). It is
really designed to test the effectiveness of tests, which is
sometimes known as Coverage.


The doctest module enables us to embed tests inside the module
documentation, and we can give it sample output which it will
check.
unittest is based upon JUnit from Java, and is a comprehensive
test framework.
Other testing regimes are available, for example 'nose' is popular
and can be downloaded from PyPI.
Of course, like anything in Python, you can always write your own.



```python
import pandas as pd
import cProfile

df = pd.DataFrame({
    'a': range(1000000),
    'b': range(1000000, 0, -1)
})

def multiply_columns():
    df['c'] = df['a'] * df['b']

cProfile.run('multiply_columns()')
```

- ncalls: the number of times the function was called
- tottime: the total time spent executing the function, excluding time spent in called functions
- percall: the average time spent per function call (tottime divided by ncalls)
- cumtime: the total time spent executing the function, including time spent in called functions
- percall: the average time spent per function call, including time spent in called functions (cumtime divided by ncalls)
- filename:lineno(function): the file name, line number, and function name of the function being profiled

```python
#unit test 
import numpy

class calculator:
    
    @staticmethod
    def multiply(number_1, number_2):
        return number_1 * number_2
    
```

```python
import unittest

class caclculator_tests(unittest.TestCase):
    def test_multiply(self):
        #given two numbers 
        number_1 = 12
        number_2 = 2
        #when they are passed to the multiply method 
        result = calculator.multiply(number_1, number_2)
        #then the product is returned 
        self.assertEquals(24, result)

unittest.main()
```

```python
import unittest

class caclculator_tests(unittest.TestCase):
    def test_multiply(self):
        #given two numbers 
        number_1 = 12
        number_2 = 2
        #when they are passed to the multiply method 
        result = calculator.multiply(number_1, number_2)
        #then the product is returned 
        self.assertEquals(24, result)
        
    def test_multiply_list(self):
        #given an array of numvers 
        array_of_numbers = [2, 10, 4]
        #when it is passed to the multiply method 
        result = calculator.multiply(array_of_numbers)
        #then the product is returned
        self.assertEquals(80, result)
     
```

```python
import numpy 
class calculator:
    
    @staticmethod
    def multiply(*obj):
        result = 0
        for value in list(obj):
            processed_value = calculator._process_multiply(value)
            if value == list(obj)[0]:
                result = processed_value
            else:
                result *= processed_value
        return result 
            
            
    @staticmethod       
    def _process_multiply(obj):
        if isinstance(obj, int):
            return obj
        elif isinstance(obj, list):
            return numpy.prod(list(obj))
        
```
