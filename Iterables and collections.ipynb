Creatinng an iterator in python. In thefollowing code the __init__() method assignes the value to it's own num. Because in this class we have defined the __iter__ and __next__ this makes this class an iterator. 
The __iter__ method fetches the iterator and initiates the iteration. the class Sequence IS an iterator so it returns itself
The __next__ method will fetch the current value from the iterator and moves it to the next state when the call to next happens. the update is incremental by 2 to update num by 2 


```python
class Sequence:
    def __init__(self):
        self.num = 1
    
    def __iter__(self):
        return self
    
    def __next__(self):
        value = self.num
        self.num +=2
        return value
```

```python
iter = Sequence()
print(next(iter))
print(next(iter))
print(next(iter))
print(next(iter))
```

A generator privides a better way to create iterators, in this case we will define a function which, instead of using a return statement, will utilise yield. With the use of yield there is some additional functionality that is made available, this  functionality retains the state of the function and will remmber what the next value is by starting at it when called again 

```python
def games():
    yield "Call of doozy"
    yield "Breath of the vile"
    yield "Splitfoo"
    yield "Among Dem"

#implement the iterator 
for thing in games():
    print(thing)
```

Clearly this is the same stype output as a for in but there is a performance enhancement, they an be compared to list comprehensions. Review this code to see a generator in action, it looks like a list comprehension 

```python
my_generator = (x**3 for x in range(10))
next(my_generator)
next(my_generator)
next(my_generator)
```

# deque
- The deque(deck) collection allows either-end-operations on a data collection 
- they can be declared as empty or can be loaded with different iterables on declaration 
- as for the namedtuple collection this needs to be imported




```python
from collections import deque

#empty deque
my_deque = deque()
print(my_deque)
```

```python
#adding iterables
my_deque = deque((3,44,5,6))
print(my_deque)
```

```python
#adding dictionary iterable 
dic = {"one": 1, "two": 2, "three": 3}

print(deque(dic.keys()))
print(deque(dic.values()))
print(deque(dic.items()))
```

```python
#efficiently popping and appending the deque
my_deque = deque([44,33,22,11])
print(my_deque)
my_deque.appendleft(99)

print(my_deque)

my_deque.append(29)

print(my_deque)

print(my_deque.popleft())
```

# Chainmap
This is another collection that is particularty more efficient at loooking things up. It's possible with chainmap to group together several dictionaries and other mappings so that they appear as one, although it is worth noting that in this case there can be multiple kney replications, a search will only return the first. 

```python
from collections import ChainMap

msisdn_dict = {'8801500121131' : 'Mr clark', '8801500123456' : 'Mr Patel'}
locations_dict = {'mumbai' : 'north west', 'nagpur' : 'central'}

chain = ChainMap(msisdn_dict, locations_dict)

print("All the ChainMap contents are ")
print(chain)

print("All the keys ")
print(list(chain.keys()))

print("All the values ")
print(list(chain.values()))
```

  It is possible to add to the chainmap

```python
new_dict = {'mumbai' : 28000000, 'nagpur' : 3900000}
chain = chain.new_child(new_dict)
print(chain)
```

```python
#Reviewing the searching of these 
print(f"mumbai in one direction this is {chain['mumbai']}")
chain.maps = reversed(chain.maps)
print(f"mumbai in the other direction is {chain['mumbai']}")
```

# Counter
Counter is a pythonic way of counting, works over items that are hashable but copes with duplication understan ding how this is used is a skill for developers. It's a good way of determining frequency of something in a dataset 

```python
from collections import Counter
import re
with open("Shakespeare.as.you.like.it.txt", 'r') as fil:
    as_list = list(fil.read().replace('\n','').replace('\t','').split(' '))
    as_set = list(set(as_list))
    #print(as_set)
    counter = {}
    pattern = '([a-zA-Z]){1,}'
    for word in as_set:
        clean_word = re.search(pattern, word)
        if clean_word != None:
            if word != '' and re.match(pattern,word):
                key_val = clean_word.group()
                try:
                    counter[key_val]+=1
                except:
                    counter[key_val] = 1
    print(counter)
```

```python
with open("Shakespeare.as.you.like.it.txt", 'r') as fil:
    as_list = list(fil.read().replace('\n','').replace('\t','').split(' '))
    counter = Counter(as_list)
    print(counter)
```

# Ordered Dictionary
Ordered dictionaries can be useful if we wanted to present the data in a certain order. For a standard dict that does not store things in a ordered way, additional code would be needed to do that process. Also worth noting that changes in the values of dict will change the position in the output but not so the case with ordered dict. Note that there is also issues with adding and removing from ordered dict as the positional data gets lost 


```python
from collections import OrderedDict
non_ordered = {"alice" : 10, "bob" : 3, "charlie" : 11 , "mongoose" : 44}
ordered_dict = OrderedDict({"alice" : 10, "bob" : 3, "charlie" : 11 , "mongoose" : 44})


for itm in non_ordered.items():
    print(itm)

print('\n\nOrdered ')

for itm in ordered_dict.items():
    print(itm)

non_ordered['alice'] = 100

print('\n\nnon Ordered ')

for itm in non_ordered.items():
    print(itm)

ordered_dict.move_to_end('alice')

print('\n\nOrdered with first element moved')

for itm in ordered_dict.items():
    print(itm)
```
